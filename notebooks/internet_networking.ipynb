# INTERNET NETWORKING GOES THROUGH THESE STAGES

when you take ur phone or pc, and browse the internet or go to a website, your pc needs the following: a number, a connection and a language.

Your pc consults the DNS Server to get the ip address,after the dns gets the ip address;

your pc starts a tcp to establish a connection with the other computer (a handshake)

your computer --> synchronize "Syn"

server --> synchronize-acknowledge "Syn-Ack"

your computer --> acknowledge "Ack"

🏁 Summary of the Stages

DNS: Get the IP address.

TCP: Establish the connection (Handshake).

HTTP Request: Ask for the page (GET).

HTTP Response: Receive the data (200 OK).

\_ _note:_ \_Python Uses the requests library to do all of it easily.

```python
import requests

response = requests.get("https://www.google.com") #make  a request to get the content of a webpage

print(response.status_code) #To know if it's 200 (Success) or 404 (Failed)

print(response.text) #This gets the raw content (usually HTML) of the page

print(response.headers) # Correct: To see metadata like date and server type
```

# 📦 The "Anatomy" of the code

1. response = requests.get("https://openai.com")
1. requests.get(...): This is the "Action." The requests toolbox goes out to the internet, does the DNS lookup, does the Handshake, and grabs the data.

1. = (The Assignment): This tells Python: "Take everything you just grabbed from the internet and save it into a variable named 'response'."

1. response: This is the name of the container. You could have named it banana or my_data, but response is the professional standard because it holds the "Response" from the server.

1. Chronology Note: You cannot use response.status_code in Step 3 until after you have created the variable; response = response.get("paste url") in the previous step.


# 🧭 The Essence of the Chronology

If we look at the life of this code, here is the "Natural Intent" of why we do it in this order:

The Intent (The get): "I want the data from that website."

The Safety Check (The status_code): "Before I try to read the data, did the server actually give it to me? If I see 200, I can proceed. If I see 404, I stop."

The Goal (The text): "Now that I know I have the data, show me the content so I can start working with it."

The Context (The headers): "I need to know what kind of data this is. Is it a webpage? Is it a JSON file? When was it last updated?"


# THE CODE UPGRADE (CODE 2)- CODE 1 WITH DEFENCE


```python
import requests
url = "https://www.google.com"
try:
  response = requests.get(url, timeout=10)
  response.raise_for_status()
  response.text 
  print("everything works fine!")
  print(response.text[:500])
except requests.exceptions.SSLError:
  print("Error: SSL handshake failed.")
except requests.exceptions.ConnectionError:
  print("Error: Could not connect to the network.")
except requests.exceptions.Timeout:
  print("Error: The request timed out.")
except requests.exceptions.HTTPError as err:
  print(f"HTTP Error occurred: {err}")
except requests.exceptions.RequestException as e:
  print(f"An unexpected networking error occurred: {e}")
```

# THE UPGRADE EXPLANATION

# 🧱 Step 1: The Library

import requests ---> DO IT FOR BOTH

# 🏗️ Step 2: The Request (The Action)

the Try Block:

Code 1: response = requests.get("https://www.google.com")

Code 2: response = requests.get(url, timeout=10)

The Upgrade:

1. We created a variable url to store the website url
2. We added timeout=10.

The Reason why we added timeout : In Code 1, if the website server is "hanging" (alive but not responding), your Python script will sit there forever, frozen.

timeout=10 tells Python: "Start a 10-second timer. If the server doesn't send data by then, kill the connection."

Full Meaning: It is a Safety Valve to prevent your program from becoming a "zombie" process that never finishes.


# 🛂 Step 3: The Success Check (The Quality Control)

Code 1: print(response.status_code)

Code 2: response.raise_for_status()

The Upgrade:

1. We replaced a "Passive Print" with an "Active Command."

The Reason: Code 1 just shows you the number (like 404). Even if it’s a failure, the code keeps running.

raise_for_status() is an Interrupt. It says: "Check the Status Code. If it is anything in the 4xx (Client Error like 404) or 5xx (Server Error like 500) range, STOP the code and jump to the error handler."

The Essence: It ensures you never try to process "bad data" or "error pages."


# 🛡️ Step 4: The Structure (The Safety Net)

Code 1: (No structure, just lines of code)

Code 2: The try: and except: blocks.

The Upgrade: We wrapped the "Action" inside a try block and created except "catchers."

The Reason: In Code 1, if your Wi-Fi is off, Python crashes with a "Traceback" (a long, scary red error message). The program dies right there.

try: Tells Python "Attempt to do this, but watch out for trouble."

except: These are the Diagnostic Nurses. If an error happens, the code doesn't die; it moves to the nurse that matches the problem.


# 🕵️ Step 5: The Diagnosis (The Specificity)

Code 1: (Gives you one generic error if it crashes)

Code 2:
except requests.exceptions.SSLError:

except requests.exceptions.ConnectionError:

The Upgrade: We are catching specific types of network failures.

1. SSLError (Secure Sockets Layer Error): This specifically tells you the "Security Bouncer" (like Zscaler or a VPN) is blocking the encrypted handshake.

2. requests.exceptions.ConnectionError: This specifically tells you the DNS (Domain Name System—the internet's phonebook) couldn't find the address, or your cable is unplugged.

3. except requests.exceptions.Timeout:
   What it does: Catches cases where the server is too slow or "hanging."
   The Scenario:
   You knocked on the door, and someone is home, but they are taking forever to open it. After your 10-second limit (the timeout=10 you set), this block triggers.
   The Result:
   Your program gives up instead of waiting forever and freezing.

4. requests.exceptions.HTTPError:
   This captures the server's specific "rejection letter" so you can print it out and see exactly which number (403, 404, etc.) it gave you.
   The Essence:
   You go from knowing "It didn't work" to knowing "Exactly why it didn't work."

5. requests.exceptions.RequestException:
   The Scenario:
   Something weird happened that doesn't fit into any other category—maybe a rare system glitch or an error we didn't name specifically.
   The Logic:
   Since this is the "Parent" of all network errors, it catches everything that leaked through the other blocks. It ensures your code has a 0% chance of showing a scary red crash screen.


# 🏁 Summary of the "Essence"

Code 1 is Optimistic: It assumes the internet is perfect, the server is fast, and your connection is always up.

Code 2 is Defensive: It assumes the internet is messy, the server might be slow (Timeout), the bouncer might be mean (SSLError), and your Wi-Fi might drop (ConnectionError).
