# LOOPS IN PYTHON


# 🔁 What is a "For Loop"?

A For Loop is a way to tell Python: "Take this list of things and do the exact same action to every single item in it, one by one, until you reach the end."

# ❓ When do we use them?

We use For Loops when we know the "range" or the "list" of things we want to process.

- To send an email to 100 customers.

- To check every file in a folder.

- To repeat an action a specific number of times.


## 🔢 1. Using Loops with Numbers

In Python, we use a helper called range() to create a list of numbers for us.

### This tells Python: "Count from 0 up to (but not including) 5"

```
for number in range(5):
    print(number)
```

How to read this in plain English:

"For every number in the range of 5, print that number."

Important Detail: Python starts counting at 0. So range(5) gives you: 0, 1, 2, 3, 4.

## 🔠 2. Using Loops with Words

Loops are great for text too! Let's take your request: writing your name 5 times.

```

```

## Wait, what is i?

In coding, i is just a traditional shorthand for "item" or "index." It’s like a temporary sticky note. Each time the loop runs, i holds the current count. Even if we don't use the sticky note inside the print() function, the loop still runs 5 times because the range is 5.

## Do I have to use "i"?

No, you absolutely do not have to use i.

In fact, you can use almost any word you want. Python doesn't care if you call it i, count, item, or even pizza.

- The "Variable" Rule
  The word you put between for and in is just a variable name that you are creating on the fly. It acts like a temporary container (a "bucket") that holds one item at a time as the loop moves through the list.

## 💡 Why do people use i then?

It’s a bit of a "hand-me-down" habit from older programming languages (like C or Fortran). In those languages, i stood for "index" or "integer." Most programmers use i when they are doing something very quick and simple. But as you get better at coding, using descriptive names is actually much better practice!

1. The range() Nuance (Assignment 5)
   You’ve used range(5) and range(1, 11) before. Here is the secret "third number" you can add: The Step.


````python
# add all the numbers between the first 50 natural numbers
total = 0
for i in range(1, 51):
  total += i
print (total)

# add all the odd numbers between the first 50 natural numbers
total2 = 0
for i in range(1, 51):
  if i % 2 == 1: #i % 2 != 0:
    total2 += i
print(total2)
````

## The continue Filter (Assignment 9)

In your assignments, you need to print 1 to 10 except 5.
Think of continue as a "teleport" button. The moment Python hits that word, it teleports back to the top of the loop for the next number, skipping everything below it for that round.


````python
for i in range(1, 11):
    if i == 5:
        continue # "Teleport" back to the top!
    print(i)
#It doesn't just skip the number; 
# it skips the rest of the instructions for that specific turn.
````

## The "Infinite Loop" Nuance (Assignment 6)

If you forget to change the condition inside the loop, it will run forever and crash your program. You must include a "counter" that moves toward the finish line.


````python
# The "Guard" condition
count = 1

while count <= 10:
    print(count)
    count = count + 1  # This is the "Finish Line" move!
````

## 🏗️ The 3 Golden Parts of a while Loop

Every successful while loop needs these three parts. If you miss one, the loop either won't start or it will never stop (the "Infinite Loop").

1. The Starting Line (Initialization)
   You have to create a variable before the loop starts so the loop has something to check.
   count = 1

2. The Gatekeeper (The Condition)
   This is the if statement's cousin. As long as this is True, the loop runs.
   while count <= 5:

3. The "Mover" (The Update)
   This is the most important part. Inside the loop, you MUST change the variable so that eventually, the Gatekeeper says "False" and lets you out.
   count = count + 1


## 🚦 How to pick > or <

Think of the condition as the "Permission to Run."

Use < (Less Than) when you are Counting Up.

Logic: "Keep going as long as the number hasn't reached the ceiling yet."

- Example:
  `while count < 10: (Start at 1, go up to 9)`.

Use > (Greater Than) when you are Counting Down.

Logic: "Keep going as long as the number is still above the floor."

- Example:
  `while countdown > 0: (Start at 10, go down to 1)`.


````python
power_level = 1

while power_level < 4:
    print(f"Charging... Level: {power_level}")
    power_level = power_level + 1

print("Fully Charged!")
````

````python
countdown = 10
while countdown > 0:
  countdown -= 1
  print(countdown)
````
