On most standard keyboards (US/UK layout), the backtick (`) is located in the top-left corner, just above the Tab key and to the left of the number 1 key.
It usually shares a key with the tilde (~).

How to find it:
On Windows/Mac (US Layout): It is the key directly below the Esc key. You don't need to hold Shift to get the backtick; just tap the key.

# ✅ formatting texts in markdown

These symbols are part of a simple language called Markdown. It’s the industry standard for programmers because it lets you format text without taking your hands off the keyboard to click buttons.

Since you're learning Python and using GitHub, you'll see this everywhere (especially in README.md files).

1. The Header System (#, ##, ###)
   The number of hashes determines the size. Think of it like an outline:

```
# Heading 1 → The Project Title

## Heading 2 → A Major Chapter

### Heading 3 → A Sub-section
```

2. Emphasis & Style (**, \*, \_)
   `Bold: Wrap text in double asterisks: **Bold Text\*\* → Bold Text`

`Italic: Wrap text in single asterisks or underscores: *Italic* or _Italic_ → Italic`

`~~Strikethrough:~~ Use double tildes: ~~Oops~~ → ~~Oops~~`

3. The "Backtick" (`) — The Programmer's Best Friend
   The backtick key is usually found right below the Esc key. It tells the computer: "Treat this text as code, not regular words."

Inline Code: Use single backticks to highlight a command or variable name inside a sentence.

Example: Type `uv sync` to see uv sync.

Code Blocks: Use triple backticks (```) to create a separate box for multiple lines of code. You can even add the language name (like python or bash) to get color coding!

4. Lists & Organization
   Bullet Points: Use a single \* or - followed by a space.

Numbered Lists: Just type 1., 2., etc.

Horizontal Rules: Type three dashes --- on a new line to create a divider line (like the one below).

5. Links & Images

- `Links: [Clickable Text](https://www.url.com)`

- `Images: ![Alt Text](https://www.image-url.com/photo.jpg)`

Pro-Tip: In Cursor or VSCode, you can create a file ending in .md (like notes.md), type your Markdown, and then press Ctrl + Shift + V to see a "Preview" of how it will actually look!

## 🧰 Part 5: Common Tips & Techniques in Cursor

### ✅ Quick Summary: Common Tasks

- **Open the Terminal:** Use `Ctrl + backtick` (the **`** key) or go to **View > Terminal**.
- **Run a Python Script:** Type `uv run your_script.py` in the terminal.
- **Add a New Package:** Use `uv add package_name` (No more `pip install`!).
- **Sync Your Environment:** Run `uv sync` to reset or update your setup.
- **Create a Jupyter Notebook:** Right-click in the file explorer > **New File** > name it `something.ipynb`.
- **Set the Notebook Kernel:** Click **"Select Kernel"** in the top-right corner of the notebook.
- **Run a Notebook Cell:** Press `Shift + Enter`.
- **Restart the Kernel:** Use the **Restart Kernel** button in the top menu of the notebook.
- **Run All Cells:** Use the **Run All** button in the top menu.
- **Adding Comments:** Use `#` for code comments in scripts, or use **Markdown cells** inside notebooks.

---

### 💡 Tips for Students

- **Restart Often:** In notebooks, use **"Restart Kernel and Run All"** frequently. This prevents "ghost" bugs caused by old variables staying in memory.
- **Save Frequently:** Even though Cursor autosaves, get into the habit of hitting `Ctrl + S` anyway.
- **Sketchpad vs. Structure:** Use **Notebooks (.ipynb)** to explore and experiment. Once your code works, move functions into a **Module (.py)** to keep things organized.
- **The "uv" Rule:** Stick to `uv add` for new packages. Only use `uv sync` if you're told to or if your environment feels "broken."

---

### 🧑‍🏫 Final Words

- **Notebooks** are your sketchpad for messy ideas.
- **Modules** are your final drafts and reusable tools.
- **uv** is your manager—it handles the heavy lifting of packages so you don't have to.
- **The Golden Rule:** If something feels weird, restart the kernel or the terminal. It solves 90% of beginner issues!
