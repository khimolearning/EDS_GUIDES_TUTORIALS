# GITIGNORE:

(.gitignore)
.gitignore is a file that contains all the files in your project u want git to ignore.
eg, you'd want git to ignore:

1. .env
2. myenv/ (your environment folder: hence the slash)
3. **pycache**/ (python junk folder )
4. .ipynb_checkpoints/ (python autosave folder)
5. .venv/ (you venv created by uv)
6. venv/

this file must be added to the parent folder as well.


# 🧹 Why these two are important:

1. \_ _pycache_ \_/: These are "compiled" Python files. They make your code run faster locally, but they are totally useless to anyone else. If you don't ignore them, your GitHub will be filled with hundreds of tiny, unreadable .pyc files.

2. .ipynb_checkpoints/: These are "autosave" files created by Jupyter Notebooks. They are temporary backups. If you upload them, it makes your repository look messy and can sometimes cause "conflicts" when you try to pull code later.

# 🛠️ The Full "Clean & Secure" Terminal Recipe

If you want to do it right in one go, here is the full sequence of commands to run in your terminal:

Bash

1. Hide the Secrets
   echo .env > .gitignore

2. Hide the Bulky Virtual Environment
   echo myenv/ >> .gitignore

3. Hide the Python "Junk"
   echo **pycache**/ >> .gitignore

4. Hide the Jupyter Autosaves
   echo .ipynb_checkpoints/ >> .gitignore
   🔍 Pro-Tip: The Single > vs Double >>
   Notice that for the first line I used > and for the others I used >>. This is a very important terminal trick:

">' (Single): This creates a new file. If the file already exists, it erases everything inside and starts fresh.

">>" (Double): This appends (adds) a new line to the bottom of the existing file without erasing what’s already there.


# 🏗️ 2. How to Start a New Project (The uv Way)

When you start your next project (let's call it "Project B"), you don't need to do anything with the old folder. You start fresh. Here is the Standard Assembly Line:

1. Create a New Folder: mkdir my-new-app and then cd my-new-app.

2. uv init: This creates your project "ID Card" (pyproject.toml) and a dummy hello.py.

3. uv venv: This creates your private "Bubble" (.venv).

4. .venv\Scripts\activate

5. uv add [library]: (e.g., uv add openai). This downloads the tool AND records it in the ID card at the same time.
