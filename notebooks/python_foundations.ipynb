```python
def even_numbers(): # No need for x here since range(50) defines it
  for x in range(50):
    if x % 2 == 0:
      print(x, "is an even number") # Use print so the loop keeps going
    else:
      print(x, "is not an even number")

even_numbers()
```

```python
def calculate_area(w, h):
  area = w * h
  return area

area_1 = calculate_area(2,5)
print("Area of the rectangle: =" , area_1)
```

```python
def leap_year(year):
  if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    return True
  else:
    return False
  

is_this_year_leap = leap_year(2026)
print(is_this_year_leap)
```
