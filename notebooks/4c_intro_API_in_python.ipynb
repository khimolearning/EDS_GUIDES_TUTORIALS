```python

import os, dotenv, openai                     #load the various libraries (toolboxes we need )
dotenv.load_dotenv(override = True)                          #load the .env file where the keys are kept 
my_secret_key = os.getenv("OPENAI_API_KEY")   #extract the api key from the envelop
client = openai.OpenAI(api_key=my_secret_key) #initialize the handshake by showing openai ur pass(api key)
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Tell me a short joke about coding."}
    ]
)                                                   #sending our message to the specific endpoint we are interested in
final_answer = response.choices[0].message.content  #extract the actual message

print(final_answer)
```

import os

from sendgrid.helpers.mail import Mail

dotenv.load_dotenv()

my_email_key = os.getenv('SENDGRID_API_KEY')

# PHASE 4: THE MESSAGE (The Content)

# ==========================================

# 'Mail' is a tool that organizes the "To", "From", and "Subject"

message = Mail(
from_email='you@example.com',

    to_emails='receiver@example.com',

    subject='Hello from Python!',

    plain_text_content='This is the body of the email.'

)

# ==========================================

# PHASE 5: THE SENDING (The Action)

# ==========================================

try: # 1. Create the SendGrid Butler (Client)

    sg = sendgrid.SendGridAPIClient(api_key=my_email_key)

    # 2. Tell the Butler to deliver the 'message' we built in Phase 4

    response = sg.send(message)

    # 3. Check if it worked (202 usually means "Accepted/Sent")

    print(f"Success! Status Code: {response.status_code}")

except Exception as e: # If the email failed (e.g., wrong email address), catch the error here

    print(f"The email failed because: {e}")


```python
import os
import dotenv
import sendgrid
from sendgrid.helpers.mail import Mail 

# Setup
dotenv.load_dotenv()
my_email_key = os.getenv('SENDGRID_API_KEY')

# Build Email
message = Mail(
    from_email='you@example.com',
    to_emails='receiver@example.com',
    subject='Hello from Python!',
    plain_text_content='This is the body of the email.'
)

# Send Email
try:
    sg = sendgrid.SendGridAPIClient(api_key=my_email_key)
    response = sg.send(message)
    print(f"Status Code: {response.status_code}")
except Exception as e:
    print(f"Error: {e}") 
```
